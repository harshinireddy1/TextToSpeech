```python
import tkinter as tk
from tkinter import *
from tkinter.ttk import Combobox
import pyttsx3
from PIL import Image, ImageTk
from tkinter import filedialog

main = Tk()
main.title("Text to Speech conversion")
main.geometry("1000x580+200+80")
main.resizable(True, True)

tts = pyttsx3.init()

def speaknow(text):
    gender=gender_box.get()
    speed=speed_box.get()
    voices=tts.getProperty('voices')
    
    def setvoice():
        if(gender == 'Male'):
            tts.setProperty('voice', voices[0].id)
            tts.say(text)
            tts.runAndWait()
        else:
            tts.setProperty('voice', voices[1].id)
            tts.say(text)
            tts.runAndWait()

    if(text):
        if(speed=='Fast'):
            tts.setProperty('rate',250)
            setvoice()
        elif(speed=='Medium'):
            tts.setProperty('rate',150)
            setvoice()
        else:
            tts.setProperty('rate',60)
            setvoice()

def load_file():
    file_path = filedialog.askopenfilename(filetypes=[("Text Files", "*.txt"), ("PDF Files", "*.pdf"), ("Word Files", "*.docx")])
    if file_path:
        if file_path.endswith('.pdf'):
            import PyPDF2
            with open(file_path, "rb") as file:
                pdf_reader = PyPDF2.PdfReader(file)
                text = ""
                for page in pdf_reader.pages:
                    text += page.extract_text()
                speaknow(text)
        elif file_path.endswith('.txt'):
            with open(file_path, "r") as file:
                text = file.read()
                speaknow(text)
        elif file_path.endswith('.docx'):
            import docx
            try:
                doc = docx.Document(file_path)
                text = ""
                for para in doc.paragraphs:
                    text += para.text + "\n"
                speaknow(text)
            except Exception as e:
                print(f"Error opening Word document: {e}")


# Load and resize the background image
bg_image = Image.open("C:/Users/harsh/Downloads/bg3.png")
bg_image = bg_image.resize((1800, 790), Image.LANCZOS)  
bg_photo = ImageTk.PhotoImage(bg_image)

# Create a Canvas widget to display the background image
canvas = Canvas(main, width=1000, height=580)
canvas.pack(fill=BOTH, expand=True)
canvas.create_image(0, 0, image=bg_photo, anchor=NW)

logo_image=PhotoImage(file = "C:/Users/harsh/Downloads/logo_image.png")
main.iconphoto(False, logo_image)

upper_frame=Frame(main, bg= "#000000" , width=1750, height=172)
upper_frame.place(x=0, y=30)

picture = PhotoImage(file = "C:/Users/harsh/Downloads/canva2.png")
Label(upper_frame, image=picture, bg="#000000").place(x=50, y=-26)

Label(upper_frame, text=" Text to Speech converter", font="Georgia 35 bold", bg="#FFF5E1", fg='black', anchor='center', width=canvas.winfo_reqwidth()).place(relx=0.5, rely=0.5, anchor='center')

text_box_width = 800  
text_box_x = (main.winfo_width() - text_box_width) // 2

text_box = Text(main, font="calibri 20", bg="white", relief=GROOVE, wrap=WORD, bd=0)
text_box.place(x=text_box_x, y=250, width=text_box_width, height=180)

def on_resize(event):
    new_text_box_x = (main.winfo_width() - text_box_width) // 2
    text_box.place(x=new_text_box_x)

main.bind("<Configure>", on_resize)

gender_box=Combobox(main, values=['Male', 'Female'], font= "Robote 12" , state= 'r', width=12)
gender_box.place(x=575,y=500)
gender_box.set('Male')

speed_box=Combobox(main, values=['Fast', 'Medium' , 'Slow'], font= "Robote 12" , state= 'r', width=12)
speed_box.place(x=775,y=500)
speed_box.set('Medium')

Label(main, text="Select Gender", font= "Georgia 15 bold", bg= "#000000" , fg="White").place(x=575,y=470)
Label(main, text="Select Speed", font= "Georgia 15 bold", bg= "#000000" , fg="White").place(x=775,y=470)

play_button= PhotoImage(file= "C:/Users/harsh/Downloads/btn2.png")
play_button = play_button.subsample(3) 

play_btn=Button(main, text= "Play", compound= LEFT, image= play_button, bg="white", width= 150,height=50 ,font="Georgia 8 bold", borderwidth='0.1c', command= lambda: speaknow(text_box.get(1.0, END)))
play_btn.place(x=655, y=600)

add_file_button = Button(main, text="Add File", bg="white", font="Georgia 10 bold", command=load_file)
add_file_button.place(x=475, y=600)

main.mainloop()
```
